#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats)

```python
engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz:3306/stats')
```

#### 4. Import the users table 

```python
connection = engine.connect()
query = 'select id, Reputation, CreationDate, displayname, lastaccessdate, websiteurl, location, aboutme, views, \
         upvotes, downvotes, accountid, age, ProfileImageUrl from users'
columnas = ['userid', 'Reputation', 'CreationDate', 'displayname', 'lastaccessdate', 'websiteurl', 'location', 'aboutme', 'views', \
         'upvotes', 'downvotes', 'accountid', 'age', 'ProfileImageUrl']
result = engine.execute(query)
```

#### 5. Rename Id column to userId

```python
pd_users = pd.DataFrame(result, columns=columnas)
pd_users
```

#### 6. Import the posts table. 

```python
queryp = 'select * from posts'
resultp = engine.execute(queryp)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
columnasp = ['postId', 'PostTypeId', 'AcceptedAnswerId', 'CreaionDate', 'Score', 'ViewCount', 'Body', 'userid', 'LasActivityDate', 'Title',  
             'Tags', 'AnswerCount', 'CommentCount', 'FavoriteCount', 'LastEditorUserId', 'LastEditDate', 'CommunityOwnedDate', 
             'ParentId', 'ClosedDate', 'OwnerDisplayName', 'LastEditorDisplayName']
pd_posts = pd.DataFrame(resultp, columns=columnasp)
pd_posts
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
pd_users_cut = pd_users[['userid', 'Reputation', 'views', 'upvotes', 'downvotes']]
pd_posts_cut = pd_posts[['postId','Score','userid','ViewCount','CommentCount']]
# pd_users_cut.shape
pd_posts_cut.head(10)
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
pd_merge = pd.merge(pd_posts_cut, pd_users_cut, on='userid', how='left')
pd_merge.shape
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
pd_merge.isnull().sum()
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
pd_sin_nulos = pd_merge[(pd_merge['userid'].isnull() == False )]
# pd_sin_nulos.shape
pd_sin_nulos = pd_sin_nulos.fillna(0)
pd_sin_nulos
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
pd_sin_nulos[['userid', 'ViewCount', 'Reputation','views','upvotes','downvotes']] = pd_sin_nulos[['userid', 'ViewCount', 'Reputation','views','upvotes','downvotes']].astype(int)
pd_sin_nulos.info()
```
