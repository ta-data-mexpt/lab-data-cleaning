#### 4. Import the users table 

```python
#Sintaxis en documentación: pd.read_sql_query(sql, con, index_col=None, coerce_float=True, params=None, parse_dates=None, chunksize=None)[source]
#Basta con el sql y el lugar de donde saca el sql
#https://pandas.pydata.org/docs/reference/api/pandas.read_sql_query.html

users_table = pd.read_sql_query ("SELECT * FROM users", mysql_engine)
users_table

#Target table:users (lo dice en el link de arriba)
```

#### 5. Rename Id column to userId

```python
users_table = users_table.rename (columns = {'Id' : 'userId'})
users_table [['userId','DisplayName']].head(3)
#users_table.head()
```

#### 6. Import the posts table. 

```python
posts_table = pd.read_sql_query ('SELECT * FROM posts', mysql_engine)
posts_table.head(3)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts_table = posts_table.rename (columns = {'Id' : 'postId','OwnerUserId':'userId'})
posts_table.head(3)
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_new_table = users_table [['userId', 'Reputation','Views','UpVotes','DownVotes']]
users_new_table.head(3)
```

```python
posts_new_table = posts_table [['postId', 'Score','userId','ViewCount','CommentCount']]
posts_new_table.head(3)
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
users_new_posts_new_merge= users_new_table.merge(posts_new_table)
users_new_posts_new_merge.head(3)
#así hace el merge automático con el nombre de la columna que encuentra igual..... 
```

```python
#pero puede ser también así para indicar cuál es el key de las dos: 
users_new_posts_new_merge2= pd.merge(users_new_table, posts_new_table, on = "userId")
users_new_posts_new_merge2.head(3)
```
