```python
def min_max_scaling(df):
    df_norm = df.copy()
    for column in df_norm._get_numeric_data():
        r = (df_norm[column].max() - df_norm[column].min()
        df_norm[column] = (df_norm[column] - df_norm[column].min()) / r)
    return df_norm    
```
